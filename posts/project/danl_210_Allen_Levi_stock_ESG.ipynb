```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

```python
url = "https://bcdanl.github.io/data/stock_history_2023.csv"
stock_history_2023 = pd.read_csv(url)
stock_history_2023
```

```python
url_2024 = "https://bcdanl.github.io/data/esg_proj_2024_data.csv"
esg_proj_2024_data = pd.read_csv(url_2024)
esg_proj_2024_data
```

```python
plt.figure(figsize=(8, 6))
```

```python
corr_1 = df_1.corr()
sns.heatmap(corr_1, annot=True, cmap='coolwarm', fmt=".2f", linewidths=.5)
```

```python
df_2 = esg_proj_2024_data[['Year', 'Total_ESG', 'Environmental', 'Social', 'Governance', 'Controversy']]
df_2
```

```python
data_1 = stock_history_2023.groupby("Symbol")["Close"].mean()
data_1 = data_1.reset_index()
data_2 = esg_proj_2024_data[["Symbol", "Environmental", "Social", "Governance", "Controversy"]]
r_data = pd.merge(data_1, data_2, on = "Symbol")
r_data = r_data.dropna()
r_data
```

```python
sns.lmplot(x='Environmental', y='Close',data=r_data)
```

```python
X = r_data[['Environmental', 'Social', 'Governance', 'Controversy']]
Y = r_data['Close']

X = sm.add_constant(X)


model = sm.OLS(Y, X).fit()
print(model.summary())
```
